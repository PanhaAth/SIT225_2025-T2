```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the cleaned CSV
df = pd.read_csv('gyroscrope_readings.csv', parse_dates=['timestamp'])
df = df.set_index('timestamp') # Set timestamp as index for plotting
```

```python
# Create a figure with 4 subplots
fig, axes = plt.subplots(4, 1, figsize=(12, 10))
fig.suptitle('Gyroscope Data Analysis', fontsize=16)
# Plot individual axes
axes[0].plot(df.index, df['gyro_x'], label='Gyro X', color='r')
axes[0].set_ylabel('Rad/s')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(df.index, df['gyro_y'], label='Gyro Y', color='g')
axes[1].set_ylabel('Rad/s')
axes[1].legend()
axes[1].grid(True)

axes[2].plot(df.index, df['gyro_z'], label='Gyro Z', color='b')
axes[2].set_ylabel('Rad/s')
axes[2].legend()
axes[2].grid(True)

# Plot all together
axes[3].plot(df.index, df['gyro_x'], label='Gyro X', color='r', alpha=0.7)
axes[3].plot(df.index, df['gyro_y'], label='Gyro Y', color='g', alpha=0.7)
axes[3].plot(df.index, df['gyro_z'], label='Gyro Z', color='b', alpha=0.7)
axes[3].set_ylabel('Rad/s')
axes[3].set_xlabel('Timestamp')
axes[3].legend()
axes[3].grid(True)

plt.tight_layout()
plt.savefig('gyroscope_plots.png', dpi=150)
plt.show()
```
