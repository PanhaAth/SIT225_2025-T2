```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

dx= pd.read_csv("Accelerometer_X.csv")
dy= pd.read_csv("Accelerometer_Y.csv")
dz= pd.read_csv("Accelerometer_Z.csv")

dx.columns = ["time", "x"]
dy.columns = ["time", "y"]
dz.columns = ["time", "z"]

```

```python
for df in (dx, dy, dz):
    df["time"] = pd.to_datetime(df["time"], errors="coerce")
    df.dropna(subset=["time", "value"], inplace=True)
    df.sort_values("time", inplace=True)

#plot X graph
plt.figure(figsize=(10, 4))
plt.plot(dx["time"], dx["value"], color = "red")
plt.title("Accelerometer X");
plt.xlabel("Time"); 
plt.ylabel("x value"); 
plt.grid(True)
plt.show()
```

```python
#plot Y graph
plt.figure(figsize=(10, 4))
plt.plot(dy["time"], dy["value"], color = "blue")
plt.title("Accelerometer Y");
plt.xlabel("Time"); 
plt.ylabel("y value"); 
plt.grid(True)
plt.show()
```

```python
#plot Z graph
plt.figure(figsize=(10, 4))
plt.plot(dz["time"], dz["value"], color = "green")
plt.title("Accelerometer Z");
plt.xlabel("Time"); 
plt.ylabel("z value"); 
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(dx["time"], dx["value"], label="X", color="red")
plt.plot(dy["time"], dy["value"], label="Y", color="green")
plt.plot(dz["time"], dz["value"], label="Z", color="blue")
plt.title("Accelerometer X, Y, Z Combined")
plt.xlabel("Time"); plt.ylabel("Acceleration"); plt.legend(); plt.grid(True)
plt.show()
```
