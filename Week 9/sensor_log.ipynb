```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sensor_log.csv", encoding="latin1", comment="#")

# Convert ms to minutes
df["time_min"] = df["timestamp_ms"] / 1000 / 60


```

```python
plt.figure(figsize =(10,5))
plt.plot(df["time_min"], df["temperature_C"], label = "Temperature °C", color = "red")
plt.xlabel("Time (minute)")
plt.ylabel("Temperature °C")
plt.title("Temperature Over Time")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["time_min"], df["humidity_percent"], label="Humidity (%)", color="blue")
plt.xlabel("Time (minutes)")
plt.ylabel("Humidity (%)")
plt.title("Humidity Over Time")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["time_min"], df["distance_cm"], label="Distance (cm)", color="green")
plt.xlabel("Time (minutes)")
plt.ylabel("Distance (cm)")
plt.title("Distance Over Time")
plt.grid(True)
plt.legend()
plt.show()
```
