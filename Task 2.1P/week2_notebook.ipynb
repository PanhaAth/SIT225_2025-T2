```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
import os

n_values = 100
y_values = []

# Create data (y_values) randomly between 1 and 100.
for i in range(n_values):
    y_values.append(random.randint(1, 100))

x_values = range(n_values)  # X is sequence of values 0-99
plt.plot(x_values, y_values)
plt.show()
```

```python
# 
# Activity 1: Create data so that the plot draws an 
# ascending line (y_values increase at any rate).
# 
x1 = np.linspace (0, 10, 100)
y1 = 2 * x1

plt.figure(figsize = (12, 4))
plt.subplot(1, 3, 1)
plt.plot(x1, y1)
plt.title("Activity 1: Ascending Line")
plt.xlabel("x values")
plt.ylabel("y values")
```

```python
# 
# Activity 2: Create data so that the plot draws a 
# descending line (y_values decrease at any rate).
# 
x2 = np.linspace(0, 10, 100)
y2 = -1.5 * x2 + 10

plt.subplot(1, 3, 2)
plt.plot(x2, y2)
plt.title("Activity 2: Descending Line")
plt.xlabel("x values")
```

```python
# 
# Activity 3: Create data so that the plot draws a 
# wave. You can consider using Python's math libarary, which has
# a sin function (detail https://www.w3schools.com/python/ref_math_sin.asp).
# 

x3 = np.linspace(0, 4 * math.pi, 100)
y3 = 3 *np.sin(x3)

plt.subplot(1, 3, 3)
plt.plot(x3, y3)
plt.title("Activity 3: Wave Pattern")
plt.xlabel("x values")

plt.tight_layout()
plt.show()
```

```python
def load_accelerometer_data():
    filepath = "accelerometer_data/accelerometer_data.csv"
    if not os.path.exists(filepath):
        raise FileNotFoundError("No accelerometer data file found")

    df = pd.read_csv(filepath, header = None, names= ['timestamp', 'x', 'y', 'z'])

    df['datetime'] = pd.to_datetime(df['timestamp'], format="%Y%m%d%H%M%S")


    start_time = df['datetime'].min()
    df['seconds'] = (df['datetime'] - start_time).dt.total_seconds()

    return df

def plot_accelerometer_data(df):
    plt.figure(figsize = (12, 8))

    plt.subplot(3, 1, 1)
    plt.plot(df['seconds'], df['x'], 'r', label = 'X-axis')
    plt.ylabel('Acceleration (m/s^2)')
    plt.title('X-axis Acceleration')
    plt.grid(True)

    plt.subplot(3, 1, 2)
    plt.plot(df['seconds'], df['y'], 'g', label = 'Y-axis')
    
    plt.subplot(3, 1, 3)
    plt.plot(df['seconds'], df['z'], 'b', label = 'Z-axis')
    plt.ylabel('Acceleration(m/s^2)')
    plt.xlabel('Time (seconds)')
    plt.title('Z-axis Acceleration')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

def plot_combined_accelerometer(df):
    plt.figure(figsize =(12, 6))
    plt.plot(df['seconds'], df['x'], 'r', label = 'X-axis')
    plt.plot(df['seconds'], df['y'], 'g', label = 'Y-axis')
    plt.plot(df['seconds'], df['z'], 'b', label = 'Z-axis')

    plt.xlabel("Time (seconds)")
    plt.ylabel("Acceleration (m/s^2)")
    plt.title("Combined Accelerometer Data")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    try:
        df = load_accelerometer_data()

        plot_accelerometer_data(df)

        plot_combined_accelerometer(df)

    except Exception as e:
        print(f"Error: {e}")
```

```python
def clean_accelerometer_data():
    filepath = "accelerometer_data/accelerometer_data.csv"
    if not os.path.exists(filepath):
        raise FileNotFoundError("No accerlerometer data file found")

    df = pd.read_csv(filepath, header = None, names= ['timestamp', 'x', 'y', 'z'])

    df_clean = df.apply(pd.to_numeric, errors = "coerce").dropna()

    df_clean = df_clean[(df_clean[['x', 'y', 'z']].abs() <20).all(axis = 1)]

    clean_filepath = "accelerometer_data/accelerometer_clean.csv"
    df_clean.to_csv(clean_filepath, index = False)
    print(f"Cleaned data saved to {clean_filepath}")

    return df_clean

if __name__ == "__main__":
    try:
        df = clean_accelerometer_data()
        df['datetime'] = pd.to_datetime(df["timestamp"], format = "%Y%m%d%H%M%S")
        start_time = df['datetime'].min()
        df['seconds'] = (df['datetime'] - start_time).dt.total_seconds()

        plot_accelerometer_data(df)
        plot_combined_accelerometer(df)

    except Exception as e:
        print(f"Error: {e}")
```
